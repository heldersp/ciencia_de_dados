# Descrição dataset e colunas

Esse é um quadro de dados sobre diamentes com 53.940 linhas e 10 variáveis:

* (price) preço em dólares americanos (\$326--\$18.823)

* (carat) peso do diamante (0,2--5,01)

* (cut) qualidade do corte (Razoável, Bom, Muito Bom, Premium, Ideal)

* (color) cor do diamante, de J (pior) a D (melhor)

* (clarity) uma medida de quão claro é o diamante (I1 (pior), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (melhor))

* x comprimento em mm (0--10,74)

* y largura em mm (0--58,9)

* z profundidade em mm (0--31,8)

* porcentagem de profundidade total = z / média (x, y) = 2 * z / (x + y) (43--79)

* largura do topo do diamante em relação ao ponto mais largo (43–95)

```python
# Primeira pergunta

# Qual é a distribuição das diferentes categorias de cortes (cut) dos diamantes no dataset?

plt.figure(figsize=(5, 3)) # Criando uma figura 5x3
sns.countplot(data=diamonds, x='cut')
plt.title('Distribuição dos cortes dos diamantes')
plt.xlabel('Corte')
plt.ylabel('Contagem')
plt.show()

# Observação: Muitas vezes confundido com a forma do diamante, o corte é, na verdade, a classificação que determina o brilho do diamante.
```

```python
# Segunda pergunta

# Como o preço dos diamantes varia com a qualidade do corte?

price_mean = diamonds.groupby('cut')['price'].mean()
price_std = diamonds.groupby('cut')['price'].std()

plt.figure(figsize=(5, 3))
price_mean.plot(kind='bar', yerr=price_std, capsize=4)
plt.title('Preço médio dos diamantes por qualidade do corte')
plt.xlabel('Corte')
plt.ylabel('Preço médio')
plt.show()
```

```python
# Terceira pergunta

#  Existe uma correlação entre o peso em quilates e o preço dos diamantes?

plt.figure(figsize=(5, 3))
sns.scatterplot(data=diamonds, x='carat', y='price')
plt.title('Correlação entre peso em quilates e preço dos diamantes')
plt.xlabel('Peso em quilates')
plt.ylabel('Preço do diamante')
plt.show()

correlation = diamonds['carat'].corr(diamonds['price'])
print(f'Correlação entre quilates e preço: {correlation}')
```

```python
# Quarta pergunta

# Como a clareza dos diamantes afeta o preço?

price_mean_clarity = diamonds.groupby('clarity')['price'].mean()
price_std_clarity = diamonds.groupby('clarity')['price'].std()

plt.figure(figsize=(5, 3))
price_mean_clarity.plot(kind='bar', yerr=price_std_clarity, capsize=2, color='skyblue')
plt.title('Preço médio dos diamantes por clareza')
plt.xlabel('Clareza')
plt.ylabel('Preço médio')
plt.show()
```

```python
# Quinta pergunta

#  Qual é a proporção de diamantes em cada categoria de cor?

plt.figure(figsize=(5, 3))
sns.countplot(data=diamonds, x='color') # Contagem de cada categoria de cor do ds diamonds
plt.title('Proporção de diamantes por cor')
plt.xlabel('Cor')
plt.ylabel('Contagem')
plt.show()
```

```python
# Sexta pergunta

# Existe uma diferença significativa no preço dos diamantes baseando-se na cor?

plt.figure(figsize=(5, 3))
price_mean_color = diamonds.groupby('color')['price'].mean()
price_mean_color.plot(kind='bar', color='skyblue')
plt.title('Preço médio dos diamantes por cor')
plt.xlabel('Quilates')
plt.ylabel('Preço médio')
plt.show()
```
